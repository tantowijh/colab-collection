# <center>**File Concate**</center>

You should create 'mylist.txt' in the directory you will execute the command and it's contain the list of your video path.
```
# this is an example
file '/path/to/file1.mp4'
file '/path/to/file2.mp4'
file '/path/to/file3.mp4'
```



````python
# @title **Concat Videos**
#@markdown *Put file location and select format extension to concat (should be the same).* <br>
#@markdown ***Note:*** *The output file will be in the same directory as File_Location.* <br>

Format_Extension = ".mp4" #@param [".mp4", ".mov", ".mkv", ".avi", ".webm", ".m4v"]
File_Location = "/content/drive/Shareddrives/05. Torrent/04. FFmpeg/recover 1" #@param {type:"string"}
Output_Name = "IG Recover I" #@param {type:"string"}

import os
import glob
import subprocess

%cd '{File_Location}'

def mp4():
  cmd = !ffmpeg -f concat -safe 0 -i mylist.txt -c copy '{Output_Name}.mp4'
def mov():
  cmd = !ffmpeg -f concat -safe 0 -i mylist.txt -c copy '{Output_Name}.mov'
def mkv():
  cmd = !ffmpeg -f concat -safe 0 -i mylist.txt -c copy '{Output_Name}.mkv'
def avi():
  cmd = !ffmpeg -f concat -safe 0 -i mylist.txt -c copy '{Output_Name}.avi'
def webm():
  cmd = !ffmpeg -f concat -safe 0 -i mylist.txt -c copy '{Output_Name}.webm'
def m4v():
  cmd = !ffmpeg -f concat -safe 0 -i mylist.txt -c copy '{Output_Name}.m4v'

if Format_Extension == ".mp4":
  mp4()
if Format_Extension == ".mov":
  mov()
if Format_Extension == ".mkv":
  mkv()
if Format_Extension == ".avi":
  avi()
if Format_Extension == ".webm":
  webm()
if Format_Extension == ".m4v":
  m4v()
````
